##Comparison of ASU and Average ASU for Each Location for class data

```python
average_ASU = Class_reading.groupby('Location Name')['asu1'].mean().reset_index()
plt.figure(figsize=(20, 6))
ax =sns.lineplot(x='Location Name', y='asu1', data=Class_reading, marker='o', label='Class Data')
sns.lineplot(x='Location Name', y='asu1', data=average_ASU, marker='o', linestyle='dashed', label='Average_ASU')
plt.xlabel('Location')
plt.ylabel('ASU')
plt.title('Comparison of ASU and Average ASU for Each Location')
for index, row in average_ASU.iterrows():
    plt.text(row['Location Name'], row['asu1'], f'Avg: {row["asu1"]:.2f}', ha='left', va='bottom')
plt.legend()
plt.grid(True)
plt.show()
```
